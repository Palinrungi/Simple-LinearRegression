# **Model Simple LinearRegression (Memprediksi berapa jarak yang dapat ditempuh berdasarkan berapa jumlah bensin)**

## Melakukan import semua library yang dibutuhkan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.model_selection as ms
import sklearn.linear_model as lm
```

## Menyiapkan Data

```python
df1 = pd.read_csv('ML DATA/bensin.csv')
df1
```

```python
df1.tail(10)
```

```python
df1.describe()
```

## Memisahkan Data Train dan Test

memisahkan data training dan test dengan ukuran 0.2 / 80 20

```python
liter = df1[['Liter']]
kilometer = df1[['Kilometer']]
X_train, X_test, y_train, y_test = ms.train_test_split(liter, kilometer, test_size=0.2, random_state=0)
print (X_train.size, X_test.size)
```

## Visualisasi

```python
plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('liter')
plt.ylabel('kilometer')
plt.title('Konsumsi Bahan Bakar')
plt.show()
```

## Menambahkan Regression Line

```python
plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('liter')
plt.ylabel('kilometer')
plt.title('Konsumsi Bahan Bakar')
x1 = np.linspace(0,45)
y1 = 4 + 7 * x1
plt.plot(x1,y1)
plt.show()
```

## Melatih Model

Latih Model Dengan Metode OLS


```python
model1=lm.LinearRegression()
model1.fit(X_train, y_train)
```

Melihat Nilai Slope

```python
model1.coef_
```

Melihat Nilai Intercept

```python
model1.intercept_
```

Visualkan dengn diagram garis dengan memasukkan angka slope dan intercept untuk formula variabel y1

```python
plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('liter')
plt.ylabel('kilometer')
plt.title('Konsumsi Bahan Bakar')
x1 = np.linspace(0,45)
y1 = 10.64 + 6.45 * x1
plt.plot(x1,y1)
plt.show()
```

## Scoring Model

Scoring menggunakan R-Square

```python
r2 = model1.score(X_test, y_test)
print (r2)
```

Melakukan Prediksi
Ternyata dengan 60 liter bensin dapat menempuh jarak sejauh 398.13 km



```python
jarak = model1.predict([[60]])
print (jarak)
```

Melakukan prediksi dari semua isi dataset

```python
prediksi = model1.predict(X_test)
print (prediksi)
```
